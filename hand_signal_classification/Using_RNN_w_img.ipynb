```python
import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import DataLoader, TensorDataset

import numpy as np
from sklearn.preprocessing import MinMaxScaler

import os, shutil
from glob import glob
import json

import numpy as np
import cv2
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


def one_hot(labels, n_classes) :
    one_hot = [0] * n_classes 
    one_hot[labels-1] = 1
    return one_hot


root_p = './data/ori_label'
folder_list = glob(root_p+'/*')
img_set = []
label_set = []
for folder in folder_list : 
    
    imgs = []
    img_list = sorted(glob(folder + '/*.jpg'), key = lambda x : int(x.split('\\')[-1].replace('.jpg','')))
    for img_path in img_list :
        img = cv2.imread(img_path)
        img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        img = img.flatten()
        imgs.append(img)
        
    img_set.extend(imgs)
    
    
    js = json.load(open(glob(folder + '/*.json')[0]))

    label = [one_hot(int(js.get('action')), n_classes=6)] * len(img_list)

    label_set.extend(label)

print(np.array(label_set).shape)
```
