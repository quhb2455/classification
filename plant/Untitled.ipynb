```python
import cv2
import os
import matplotlib.pyplot as plt

```

```python
img0 = cv2.imread('./data/train_imgs/10005.jpg')
img1 = cv2.imread('./data/train_imgs/10000.jpg')
img2 = cv2.imread('./data/train_imgs/10002.jpg')
img3 = cv2.imread('./data/train_imgs/10015.jpg')
img4 = cv2.imread('./data/train_imgs/10003.jpg')
img5 = cv2.imread('./data/train_imgs/10009.jpg')
img6 = cv2.imread('./data/train_imgs/10017.jpg')
```

```python
fig = plt.figure()
rows = 2
cols = 4

ax0 = fig.add_subplot(rows, cols, 1)
ax0.imshow(cv2.cvtColor(img0, cv2.COLOR_BGR2RGB))
ax0.set_title('class : 0')
ax0.axis("off")

ax1 = fig.add_subplot(rows, cols, 2)
ax1.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax1.set_title('class : 1')
ax1.axis("off")

ax2 = fig.add_subplot(rows, cols, 3)
ax2.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax2.set_title('class : 2')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 4)
ax3.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
ax3.set_title('class : 3')
ax3.axis("off")

ax4 = fig.add_subplot(rows, cols, 5)
ax4.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
ax4.set_title('class : 4')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 6)
ax5.imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))
ax5.set_title('class : 5')
ax5.axis("off")

ax6 = fig.add_subplot(rows, cols, 7)
ax6.imshow(cv2.cvtColor(img6, cv2.COLOR_BGR2RGB))
ax6.set_title('class : 6')
ax6.axis("off")
plt.show()
```
