```python
def get_vector(points, x_y_z=[16, 16, 16]):
        # 3D Points -> [16,16,16]
        
        # points에서 최대 최소 값 구함
        xyzmin = np.min(points, axis=0) - 0.001
        xyzmax = np.max(points, axis=0) + 0.001
        print("xyzmin ", xyzmin)
        print("xyzmax ", xyzmax)
        

        diff = max(xyzmax-xyzmin) - (xyzmax-xyzmin)
        print(max(xyzmax-xyzmin))
        print(xyzmax-xyzmin)
        print("diff ", diff)
        xyzmin = xyzmin - diff / 2
        xyzmax = xyzmax + diff / 2
        print("xyzmin ", xyzmin)
        print("xyzmax ", xyzmax)
        
        segments = []
        shape = []

        for i in range(3):
            # note the +1 in num 
            if type(x_y_z[i]) is not int:
                raise TypeError("x_y_z[{}] must be int".format(i))
            # x, y, z 각각에 대하여 최소값 ~ 최대값까지 총 17개의 값을 만들어냄
            # ex) x ==>  -0.5 , -0.4 , -0.37 , ...., 0.4, 0.5 총 17개가 나옴
            # step : 값의 간격
            # s : 17개의 값
            s, step = np.linspace(xyzmin[i], xyzmax[i], num=(x_y_z[i] + 1), retstep=True)
            print("s ",s)
            print("step ",step)
            segments.append(s)
            shape.append(step)
        print()
        print()
        print("segments\n",segments)
        print("x - segments[0]",segments[0].shape)
        print("y - segments[1]",segments[1].shape)
        print("z - segments[2]",segments[2].shape)
        print("shape\n",shape)
        
        # voxel의 개수 
        n_voxels = x_y_z[0] * x_y_z[1] * x_y_z[2]
        n_x = x_y_z[0]
        n_y = x_y_z[1]
        n_z = x_y_z[2]

        # points에 있는 값들을 17개 값들 중 어느위치에 들어가는지 이진탐색을함
        # np.searchsorted는 이진 탐색을 해서 두번 째 파라미터가 첫번 째 파라미터인 list에서 
        # 몇번 째 index에 들어가는지를 return함
        structure = np.zeros((len(points), 4), dtype=int)
        structure[:,0] = np.searchsorted(segments[0], points[:,0]) - 1
        print("structure[:,0] ", structure[:,0])
        structure[:,1] = np.searchsorted(segments[1], points[:,1]) - 1
        print("structure[:,1] ", structure[:,1])
        structure[:,2] = np.searchsorted(segments[2], points[:,2]) - 1
        print("structure[:,2] ", structure[:,2])

        # i = ((y * n_x) + x) + (z * (n_x * n_y))
        structure[:,3] = ((structure[:,1] * n_x) + structure[:,0]) + (structure[:,2] * (n_x * n_y)) 
        print("structure[:,3]", structure[:,3], len(structure[:, :3]))
        
        vector = np.zeros(n_voxels)
        print("vector", vector.shape)
        count = np.bincount(structure[:,3])
        print("count", count, max(count), count.shape)
        vector[:len(count)] = count
        print("vector", vector)
        vector = vector.reshape(n_z, n_y, n_x)
        return vector
```
