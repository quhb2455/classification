```python
import torch
from torchvision import transforms

import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
torchvision_transforms = transforms.Compose([
    transforms.Resize((300, 300)),
    transforms.RandomCrop(224),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    transforms.RandomHorizontalFlip(p=1),
#     transfomrs.ToTensor(),
])
```

```python
fig = plt.figure(figsize=(15, 15))
rows = 1
cols = 2


img_1 = Image.open('1.jpg').convert('RGB')
fig.add_subplot(rows, cols, 1)
plt.imshow(img_1)

img_2 = Image.open('2.jpg').convert('RGB')
fig.add_subplot(rows, cols, 2)
plt.imshow(img_2)
```

```python
img_1_trans = torchvision_transforms(img_1)
img_2_trans = torchvision_transforms(img_2)

fig = plt.figure(figsize=(15, 15))
rows = 1
cols = 2

fig.add_subplot(rows, cols, 1)
plt.imshow(img_1_trans)


fig.add_subplot(rows, cols, 2)
plt.imshow(img_2_trans)

```

```python
import torch.nn as nn


pred = torch.tensor([[0.1, 0.2, 0.7],
                    [0.5, 0.1, 0.4]])
gt = torch.tensor([2, 1])

gamma = 2
alpha = 1

ce_loss = nn.CrossEntropyLoss(reduction='none')(pred, gt)
display(ce_loss)
pt = torch.exp(-ce_loss)
focal_loss = alpha * ((1 - pt) ** gamma) * ce_loss
display(focal_loss)
display(torch.mean(focal_loss))
display(torch.sum(focal_loss))
```
