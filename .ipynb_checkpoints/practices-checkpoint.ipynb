```python
def parse_function_gaussian(filename, label) :
    print(filename)
    print(label)
    img_string = tf.io.read_file(filename)
    img_decoded = tf.image.decode_jpeg(img_string, channels=3)
#     img = cv2.imread(filename)
#     print(img_decoded)
#     img_th = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,2)
    img_resize = tf.image.resize(img_decoded, [224, 224])

    return img_resize, label

```

```python
def create_dataset_gaussian(filenames, labels, is_training=True) :
    BATCH_SIZE = 16
    AUTOTUNE = tf.data.experimental.AUTOTUNE
    SHUFFLE_BUFFER_SIZE = 1024
    
    dataset = tf.data.Dataset.from_tensor_slices((filenames, labels))
    dataset = dataset.map(parse_function_gaussian, num_parallel_calls=AUTOTUNE)
    for img, label in dataset :
        print("label : ", label)
        cv2.imshow('image', img.numpy())
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    dataset = dataset.batch(BATCH_SIZE)
    dataset = dataset.prefetch(buffer_size=AUTOTUNE)
    
    return dataset
```

```python
def parse_function(filename, label) :
    img_string = tf.io.read_file(filename)
    img_decoded = tf.image.decode_jpeg(img_string, channels=3)
    img_resize = tf.image.resize(img_decoded, [224, 224])
    return img_resize, label
```

```python
def create_dataset(filenames, labels, is_training=True) :
    BATCH_SIZE = 16
    AUTOTUNE = tf.data.experimental.AUTOTUNE
    SHUFFLE_BUFFER_SIZE = 1024
    
    dataset = tf.data.Dataset.from_tensor_slices((filenames, labels))
    dataset = dataset.map(parse_function, num_parallel_calls=AUTOTUNE)

    dataset = dataset.batch(BATCH_SIZE)
    dataset = dataset.prefetch(buffer_size=AUTOTUNE)
    
    return dataset
```

```python
import os

imgs = []
names = []
labels = []

with open('./casia/label/label.txt', 'r') as f :
    lines = f.readlines()
    for line in lines :
        line = line.strip()
        names.append(line.split(' ')[0])
        labels.append([int(label) for label in line.split(' ')[1:]])
    
# for img in os.listdir('./casia/img/generate_test') :
#     imgs.append(os.path.join('./casia/img/generate_test', img))
for img in names :
    imgs.append(os.path.join('./casia/img/generate_test', img))

dataset = create_dataset(imgs, labels)
```

```python
import tensorflow_hub as hub

feature_extractor_url = "https://tfhub.dev/google/imagenet/mobilenet_v2_100_224/feature_vector/4"
feature_extractor_layer = hub.KerasLayer(feature_extractor_url,
                                         input_shape=(224,224,3))
feature_extractor_layer.trainable = False
```

```python
from keras import Sequential,layers

model = Sequential()
# model.add(feature_extractor_layer)
model.add(layers.Input(shape=(224,224,3)))
model.add(layers.Conv2D(16, 3, padding='same', activation='relu'))
model.add(layers.MaxPool2D())
model.add(layers.Conv2D(16, 3, padding='same', activation='relu'))
model.add(layers.MaxPool2D())
model.add(layers.Conv2D(16, 3, padding='same', activation='relu'))
model.add(layers.MaxPool2D())
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(6, activation='sigmoid'))

model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
import os

def parse_data(filename, label) :
    img_read = tf.io.read_file(filename)
    img_decode = tf.image.decode_jpeg(img_read)
    img = tf.image.resize(img_decode, [224,224])
    img = tf.image.grayscale_to_rgb(img)

    return img, label

def create_dataset(filenames, labels, batch_size) :
    dataset = tf.data.Dataset.from_tensor_slices((filenames,labels))
    dataset = dataset.map(parse_data, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
    
    return dataset

def read_labels(label_txt) :
    name = []
    label = []
    with open(label_txt, 'r') as f :
        lines = f.readlines()
        for line in lines :
            contents = line.split()
            name.append(contents[0])
            label.append([int(i) for i in contents[1:]])
    return name, label

def read_imageList(path, names) :
    return [os.path.join(path, name) for name in names]




# def parse_data2(filename) :
#     img_read = tf.io.read_file(filename)
#     img_decode = tf.image.decode_jpeg(img_read)
#     img = tf.image.resize(img_decode, [224,224])

#     return img



batch_size = 16
image_path = './casia/img'

names, labels = read_labels('./casia/label/label.txt')
images = read_imageList(image_path, names)
dataset = create_dataset(images, labels, batch_size)

# for image in images :
#     img = parse_data2(image)
#     if img.shape[2] == 1:
#         print(image, img.shape)

# for i,name in enumerate(names) :
#     if i > 288 :
#         print(i)
#         print(name)
#     if i > 305 :
#         break
for i,(x, name)in enumerate(zip(dataset, names)) :
    print(i)
    print(name)
    print(x)
    
```

# dataset w/ augmentation 불러오기

```python
from tqdm import tqdm 
import os, cv2
import numpy as np

def read_labels(label_txt) :
    names = []
    labels = []
    with open(label_txt, 'r') as f :
        lines = f.readlines()
        for line in lines :
            contents = line.split()
            names.append(contents[0])
            labels.append([int(i) for i in contents[1:]])
    return names, np.array(labels)

def read_imageList(path, names) :
    return [os.path.join(path, name) for name in names]

def create_training_img_set(filenames) :
    imgs = []
    
    for filename in tqdm(filenames) :
        img_read = cv2.imread(filename)
        img_rgb = cv2.cvtColor(img_read, cv2.COLOR_BGR2RGB)
        imgs.append(cv2.resize(img_rgb, dsize=(224,224)))
    
    return np.array(imgs)
    


batch_size = 16
image_path = './casia/img'

names, labels = read_labels('./casia/label/label.txt')
filenames = read_imageList(image_path, names)
train_img_set = create_training_img_set(filenames)
```

```python
print("training img set shape :",train_img_set.shape)
print("training label set shape :",labels.shape)
```

# Define metrics 

```python
metrics = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Accuracy(name='acc'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall')
]
```

# Define Callbacks

```python
monitor = 'val_loss'
ckpt_path = './abc/test.h5'

# os.makedirs(ckpt_path,exist_ok=True)

callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor=monitor, patience=3, restore_best_weights=True ),
    tf.keras.callbacks.ModelCheckpoint(ckpt_path, monitor=monitor,
                                       verbose=0, 
                                      save_weights_only=False,
                                      mode='auto',
                                      save_freq='epoch'),
    tf.keras.callbacks.ReduceLROnPlateau(monitor=monitor, factor=0.1,
                                        patience=5, min_lr=1e-13)
]
```
