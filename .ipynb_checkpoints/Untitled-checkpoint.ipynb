```python
import cv2, os
import matplotlib.pyplot as plt

img = cv2.imread('./1.jpg')
print(img.shape)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
first_img = img[:, :224, :]
second_img = img[:, 224:224*2, :]
thrid_img = img[:, 224*2:224*3,:]

print(first_img.shape)
print(second_img.shape)
print(thrid_img.shape)
```

```python
fig = plt.figure()
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(cv2.cvtColor(first_img, cv2.COLOR_BGR2RGB))
ax1.set_title('first_img')
ax1.axis("off")
 
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(cv2.cvtColor(second_img, cv2.COLOR_BGR2RGB))
ax2.set_title('second_img')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(cv2.cvtColor(thrid_img, cv2.COLOR_BGR2RGB))
ax3.set_title('thrid_img')
ax3.axis("off")

plt.show()
```

```python
color = [255,0,0]
cnt = 0
imgs = [first_img, second_img, thrid_img]
with open('1.txt', 'r') as f :
    for line in f.readlines() :
        if line[:-1] == '68' :
            cnt += 1
            continue
        
        if cnt != 0  :
            x_y =line.split(',')
            x = round(float(x_y[0]))
            y = round(float(x_y[1]))
            
            if cnt == 1 :
                imgs[2] = cv2.circle(imgs[2], (x, y), 1, color, 1)
            if cnt == 2 :
                imgs[0] = cv2.circle(imgs[0], (x, y), 1, color, 1)
            if cnt == 3 :
                imgs[1] = cv2.circle(imgs[1], (x, y), 1, color, 1)
```

```python
fig = plt.figure()
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(cv2.cvtColor(first_img, cv2.COLOR_BGR2RGB))
ax1.set_title('first_img')
ax1.axis("off")
 
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(cv2.cvtColor(second_img, cv2.COLOR_BGR2RGB))
ax2.set_title('second_img')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(cv2.cvtColor(thrid_img, cv2.COLOR_BGR2RGB))
ax3.set_title('thrid_img')
ax3.axis("off")

plt.show()
```

```python
cv2.imwrite('lm_1.jpg',first_img)
cv2.imwrite('lm_2.jpg',second_img)
cv2.imwrite('lm_3.jpg',thrid_img)
```
