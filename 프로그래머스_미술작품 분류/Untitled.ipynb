```python
from glob import glob

base_path = "./data/train"

classes = glob(base_path+'/*')

total = 0
for class_name in classes :
    imgs = glob(class_name + '/*')
    name = class_name.split('\\')[-1]
    print(name + " : " + str(len(imgs)))
    total += len(imgs)
print("--------------")
print("total : ", total)
```
