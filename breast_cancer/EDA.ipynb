```python
import matplotlib.pyplot as plt
import cv2
import os
from glob import glob
from tqdm import tqdm
import numpy as np
import pandas as pd
```

```python
img_list = glob('./data/train_imgs/*.png')
print("이미지 개수 : ",len(img_list))
```

## 이미지 가로 세로 분포

```python
width = []
height = []
for img_path in tqdm(img_list) :
    img = cv2.imread(img_path)
    height.append(img.shape[0])
    width.append(img.shape[1])
```

```python
f, axs = plt.subplots(2,1, figsize=(20,40))
axs = axs.flatten()

w_n, w_b = np.histogram(np.array(width), bins=np.arange(0, max(width), 1))
axs[0].hist(np.array(width), w_b)
axs[0].set_title('Width')
h_n, h_b = np.histogram(np.array(height), bins=np.arange(0, max(height), 1))
axs[1].hist(np.array(height), h_b)
axs[1].set_title('Height')

plt.show()
```
